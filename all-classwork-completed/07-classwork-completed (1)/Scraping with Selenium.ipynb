### Import selenium

```python
from selenium import webdriver
```

### Bring a new Chrome to life!

```python
driver = webdriver.Chrome()
```

```python
driver.get("http://www.hm.com/us/products/sale/home")
```

```python
titles = driver.find_elements_by_class_name('product-title')
for title in titles:
    print(title.text)
```

```python
urls = driver.find_elements_by_class_name('product-url')
for url in urls:
    print(url.get_attribute('href'))
```

```python
# If you want to see the page source code instead of using the inspector
# driver.page_source
```

```python
images = driver.find_elements_by_class_name('product-image')
for image in images:
    print(image.get_attribute('src'))
```

```python
prices = driver.find_elements_by_class_name('product-price')
for price in prices:
    print(price.text)
```

```python
# Find all of the groups of content on the page,
# The things that are going to become our rows
# in this case, it's each product box
products = driver.find_elements_by_class_name('product-list-item')

# Then for every one of those elements...
for product in products:
    # Create and empty dictionary
    row = {}
    
    # This helps things be visually broken up
    print('-----------')
    
    # Find the title and the price etc
    # be sure to use .find_element and NOT .find_elements
    title = product.find_element_by_class_name('product-title')
    # print(title.text)
    # Store the title in our new dictionary
    row['title'] = title.text
    
    price = product.find_element_by_class_name('product-price')
    # print(price.text)
    row['price'] = price.text
    
    url = product.find_element_by_class_name('product-url')
    # print(url.get_attribute('href'))
    row['url'] = url.get_attribute('href')
    
    print("Dictionary looks like", row)
    # my_dictionary = {
    #     'url': url.get_attribute('href'),
    #     'title': title.text,
    #    'price': price.text
    # }
    
    
```

```python
# Find all of the groups of content on the page,
# The things that are going to become our rows
# in this case, it's each product box
products = driver.find_elements_by_class_name('product-list-item')

rows = []

# Then for every one of those elements...
for product in products:
    print('-----------')
    # Create and empty dictionary
    row = {}
    
    # Find the title and the price etc
    # be sure to use .find_element and NOT .find_elements
    # Store the title in our new dictionary
    title = product.find_element_by_class_name('product-title')
    row['title'] = title.text
    
    price = product.find_element_by_class_name('product-price')
    row['price'] = price.text
    
    url = product.find_element_by_class_name('product-url')
    row['url'] = url.get_attribute('href')
    
    print("Dictionary looks like", row)
    
    # Add that dictionary to the list of rows
    rows.append(row)
    
    
```

```python
import pandas as pd

df = pd.DataFrame(rows)
df.head(10)
```

```python
# Save it as a CSV
df.to_csv("sales.csv", index=False)
```
